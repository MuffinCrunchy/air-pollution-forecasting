# **Import Library**

```python
import os
import numpy as np
import pandas as pd
```

# **Upload Dataset**

```python
raw_datasets_directory = "/content/air-pollution-forecasting/datasets/raw-datasets/"

datasets = []
for dataset in os.listdir(raw_datasets_directory):
    if dataset.endswith(".csv"):
        print("Memuat file {0}...".format(dataset))
        datasets.append(pd.read_csv(os.path.join(raw_datasets_directory, dataset)))

merged_csv = pd.concat(datasets, axis=0)
merged_csv.to_csv("jkt-ispu-2011-2021-raw.csv", index=False)
print("Merge completed...")
```

```python
df = pd.read_csv('jkt-ispu-2011-2021-raw.csv')
df = df.sort_values(["tanggal"], ascending=True)
df
```

```python
df.info()
```

**Delete Column**

```python
df.drop(df.columns[range(14, 25)], axis=1, inplace=True)
df.drop(['pm25', 'max', 'critical', 'categori', 'kategori', 'keterangan'], axis=1, inplace=True)
df.info()
```

# **Clean Dataset**

```python
df['s02'].unique()
```

```python
df['so2'] = df['so2'].fillna(df['s02'])
df.drop('s02', axis=1, inplace=True)
df.info()
```

```python
df = df.dropna(subset=['tanggal'])
df
```

# **Rearrange Dataset**

```python
col_ord = ['tanggal', 'stasiun', 'pm10', 'so2', 'co', 'o3', 'no2']
df = df[col_ord]
df.info()
```

```python
df
```

```python
df['tanggal'] = pd.to_datetime(df.tanggal)
df['tanggal'] = df['tanggal'].dt.strftime('%Y-%m-%d')
df
```

```python
df['stasiun'].unique()
```

```python
df.replace('DKI1 (Bunderan HI)','Jakarta Pusat', inplace=True)
df.replace('DKI1','Jakarta Pusat', inplace=True)
df.replace('DKI2 (Kelapa Gading)','Jakarta Utara', inplace=True)
df.replace('DKI2','Jakarta Utara', inplace=True)
df.replace('DKI3 (Jagakarsa)', 'Jakarta Selatan', inplace=True)
df.replace('DKI3', 'Jakarta Selatan', inplace=True)
df.replace('DKI4 (Lubang Buaya)','Jakarta Timur', inplace=True)
df.replace('DKI4','Jakarta Timur', inplace=True)
df.replace('DKI5 (Kebon Jeruk)','Jakarta Barat', inplace=True)
df.replace('DKI5','Jakarta Barat', inplace=True)
df.replace('DKI5 (Kebon Jeruk) Jakarta Barat','Jakarta Barat', inplace=True)
df['stasiun'].unique()
```

```python
df
```

# **Save as New Datasets**

```python
df = df.replace('---', np.NaN)
df
```

```python
df = df.sort_values(["tanggal"], ascending=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv("jkt-ispu-2011-2021.csv", index=False)
```
