## **Plotting**

```python
data_17_s = go.Scatter(
    x = df_dates_s['tanggal'],
    y = df_17_s['co'],
    mode = 'lines',
    name='2017'
)
data_18_s = go.Scatter(
    x = df_dates_s['tanggal'],
    y = df_18_s['co'],
    mode = 'lines',
    name='2018'
)
data_19_s = go.Scatter(
    x = df_dates_s['tanggal'],
    y = df_19_s['co'],
    mode = 'lines',
    name='2019'
)
data_20_s = go.Scatter(
    x = df_dates_s['tanggal'],
    y = df_20_s['co'],
    mode = 'lines',
    name='2020'
)
data_21_s = go.Scatter(
    x = df_dates_s['tanggal'],
    y = df_21_s['co'],
    mode = 'lines',
    name='2021'
)
layout = go.Layout(
    title = "Air Pollution (Summer)",
    xaxis = {'title' : "Date"},
    yaxis = {'title' : "Carbon Monoxide(CO)"}
)

fig = go.Figure(data=[data_17_s, data_18_s, data_19_s, data_20_s, data_21_s], layout=layout)
fig.show()
```
